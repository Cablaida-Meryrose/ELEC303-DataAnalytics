<h1> BASIC OF MACHINE LEARNING </h1>

```python
import pandas as pd 
import warnings
warnings.filterwarnings("ignore", message="X does not have valid feature names")
```

```python
df = pd.read_csv('Titanic.csv') 
```

```python
df
```

```python
# Drop irrelevant columns
df = df[["Survived", "Pclass", "Sex", "Age","Embarked"]]

# Drop missing values
df = df.dropna()

# Encode categorical variable
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
df["Embarked"] = df["Embarked"].map({"C":0, "Q":1, "S":2})
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = df[["Pclass", "Sex", "Age", "Embarked"]]
y = df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

<h1> Accuracy Test </h1>

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pandas as pd

# Load and clean dataset
df = pd.read_csv("Titanic.csv")
df = df[["Survived", "Pclass", "Sex", "Age", "Fare", "Embarked"]]
df = df.dropna()

# Encode categorical variables
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
df["Embarked"] = df["Embarked"].map({"S": 0, "C": 1, "Q": 2})

```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Define features and target
X = df[["Pclass", "Sex", "Age", "Fare", "Embarked"]]
y = df["Survived"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Train model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Evaluate model
y_pred = model.predict(X_test)
print("Model Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Manual input: [Pclass, Sex, Age, Fare, Embarked]
manual_inputs = [
    [3, 0, 22, 7.25, 0],  # Lower-class male, young, low fare, Southampton
    [1, 1, 30, 80.0, 1],  # First-class female, adult, high fare, Cherbourg
]

# Predict and interpret
for passenger in manual_inputs:
    prediction = model.predict([passenger])[0]
    if prediction == 1:
        print("Passenger likely to survive")
    else:
        print("Passenger unlikely to survive")
```
